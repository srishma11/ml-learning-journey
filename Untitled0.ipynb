```python
ages = [22, 15, 35, 8, 45]
for age in ages:
    if age >= 18:
        print(age, "is an adult")
    else:
        print(age, "is a minor")
```

```python
survived = [1, 0, 1, 1, 0]
for result in survived:
    if result == 1:
       print("survived")
    else:
       print("did not survive")
```

```python
for result in df['Survived'][:10]:
    if result == 1:
       print("survived")
    else:
       print("did not survive")
```

```python
count = 0
for x in df['Survived'][:10]:
    if x == 1:
       count = count + 1

print("total survived:", count)
```
